### **Data Exploration**

**Summary of attributes most/least desirable in male and female partners based on ratings.**
- Let's focus on partner ratings by gender
- Gender: 0 = female, 1 = male (according to dataset documentation)
- Let's compare how males and females rate the importance of traits in partners

```python
# Ratings given by participants to their partners
rating_columns = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']

# Filter by gender
male_ratings = df[df['gender'] == 1][rating_columns]
female_ratings = df[df['gender'] == 0][rating_columns]

# Calculate mean ratings
avg_male_ratings = male_ratings.mean().rename("Rated by Females (on M)")
avg_female_ratings = female_ratings.mean().rename("Rated by Males (on F)")

# Combine for comparison
ratings_comparison = pd.concat([avg_male_ratings, avg_female_ratings], axis=1)

# Plotting
ax = ratings_comparison.plot(kind='bar', figsize=(10, 6), title="Average Ratings of Partner Traits by Gender")
plt.ylabel("Average Rating (1-10)")
plt.xlabel("Trait")
plt.xticks(rotation=45)
plt.tight_layout()

# Add data labels
for container in ax.containers:
    ax.bar_label(container, labels=[f'{val:.2f}%' for val in container.datavalues], label_type='edge',fontsize=7, padding=3)

ratings_comparison
```

- Men value attractiveness more in a partner and they rated women higher in attractiveness than women rated men.
- Women rated men higher in intelligence and ambition than men rated women.
- Traits like sincerity and fun were rated relatively high by both genders.
- Shared interests had the lowest average ratings across the board, hinting they may be less important in initial attraction.

- Females consistently assign more importance to ambition and shared interests than males.
- Fun and sincerity are moderately valued across both groups.

These stated preferences highlight different dating priorities, especially between appearance (men) vs. personality traits (women).

Let's dive into whether shared interests or shared racial background are more influential in getting a match.

**Shared interests vs. shared racial background**

```python
# Shared interests vs. shared racial background
# 'match' = 1 if there was mutual interest
# 'samerace' = 1 if partners were of same race
# 'int_corr' = interest correlation (measure of shared interests)

# Clean up and drop NaNs relevant to this analysis
match_data = df[['match', 'samerace', 'int_corr']].dropna()

# Group by whether there's a match and calculate averages
grouped_means = match_data.groupby('match').mean()

# Plotting
ax = grouped_means.plot(kind='bar', figsize=(8, 6), title="Shared Interests and Race by Match Outcome")
plt.ylabel("Average Value")
plt.xlabel("Match (0 = No, 1 = Yes)")
plt.xticks(rotation=0)
plt.tight_layout()

# Add data labels
for container in ax.containers:
    ax.bar_label(container, labels=[f'{val:.2f}%' for val in container.datavalues], label_type='edge',fontsize=7, padding=3)

grouped_means
```

- People who matched were slightly more likely to share the same race, but the difference is small.
- Shared interests had a slightly stronger impact than racial background.
- Overall, neither factor had a dramatic effect, but shared interests edge out race as a predictor of a mutual match.

Now let's explore if being someone's first or last date of the night affects the chance of a second date.

**Speed dating order**

```python
# We use the 'order' column to determine the order of the date during the night
# Lower values = earlier in the evening, higher = later
# Let's analyze the average match rate based on order buckets (e.g., early, middle, late)

# Create buckets: early (1-3), middle (4-6), late (7+)
df_order = df[['match', 'order']].dropna()
df_order['order_bucket'] = pd.cut(df_order['order'], bins=[0, 3, 6, df_order['order'].max()], labels=['Early', 'Middle', 'Late'])

# Calculate match rate per bucket
order_match_rate = df_order.groupby('order_bucket')['match'].mean()

# Plotting
ax = order_match_rate.plot(kind='bar', figsize=(7, 5), title="Match Rate by Time of Night (Order of Date)")
plt.ylabel("Match Rate")
plt.xlabel("Order Bucket")
plt.xticks(rotation=0)
plt.tight_layout()

# Add data labels
for container in ax.containers:
    ax.bar_label(container, labels=[f'{val:.2f}%' for val in container.datavalues], label_type='edge',fontsize=7, padding=3)

order_match_rate
```

- Earlier dates have the highest match rate, with the likelihood of matching decreasing as the night progresses.
- This could be due to fatigue, comparison bias, or simply better energy early on.

**Comparison of participants' self-ratings to how their partners rated them**
