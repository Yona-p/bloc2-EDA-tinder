### **Data Exploration**

**Summary of attributes most/least desirable in male and female partners based on ratings.**
- Let's focus on partner ratings by gender
- Gender: 0 = female, 1 = male (according to dataset documentation)
- Let's compare how males and females rate the importance of traits in partners

```python
# Ratings given by participants to their partners
rating_columns = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']

# Filter by gender
male_ratings = df[df['gender'] == 1][rating_columns]
female_ratings = df[df['gender'] == 0][rating_columns]

# Calculate mean ratings
avg_male_ratings = male_ratings.mean().rename("Rated by Females (on Males)")
avg_male_ratings = female_ratings.mean().rename("Rated by Males (on Females)")

# Combine for comparison
ratings_comparison = pd.concat([avg_male_ratings, avg_male_ratings], axis=1)

# Chart
ratings_comparison.plot(kind='bar', figsize=(10, 6), title="Average Ratings of Partner Traits by Gender")
plt.ylabel("Average Rating (1-10)")
plt.xlabel("Trait")
plt.xticks(rotation=45)
plt.tight_layout()


ratings_comparison
```

- Men value attractiveness more in a partner and they rated women higher in attractiveness than women rated men.
- Women rated men higher in intelligence and ambition than men rated women.
- Traits like sincerity and fun were rated relatively high by both genders.
- Shared interests had the lowest average ratings across the board, hinting they may be less important in initial attraction.

**Comparaison of traits vs real impact**

```python
# Importance of each trait in a partner
importance_columns  = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']
importance_labels  = ['Attractiveness', 'Sincerity', 'Intelligence', 'Fun', 'Ambition', 'Shared Interests']

# Average of importance by gender
male_importance = df[df['gender'] == 1][importance_columns].mean().rename("Importance by Males")
female_importance = df[df['gender'] == 0][importance_columns].mean().rename("Importance by Females")

# Rename indices
male_importance.index = importance_labels
female_importance.index = importance_labels

# Combine into a single dataframe
importance_comparison = pd.concat([male_importance, female_importance], axis=1)

# Plotting 
importance_comparison.plot(kind='bar', figsize=(10, 6), title="Stated Importance of Partner Traits by Gender")
plt.ylabel("Average Importance (out of 100%)")
plt.xlabel("Trait")
plt.xticks(rotation=45)
plt.tight_layout()

importance_comparison
```

- For males, attractiveness is most important(27%), while women place more importance on intelligence and sincerity.
- Females consistently assign more importance to ambition and shared interests than males.
- Fun and sincerity are moderately valued across both groups.

**Correlation stated preferences with actual ratings their dates**

```python
# Ensure both sets of columns are aligned by trait
# We'll calculate correlation between each individual's stated importance and how they rated their partner

# First, filter rows with complete data
valid_rows = df[importance_columns + rating_columns].dropna()

# Now compute correlations between the stated importance and actual ratings for each trait
correlations = {
    trait: valid_rows[imp].corr(valid_rows[act])
    for trait, imp, act in zip(importance_labels, importance_columns, rating_columns)
}

# Convert to Series
correlation_series = pd.Series(correlations)

# Plotting
correlation_series.plot(kind='bar', figsize=(10, 6), title="Correlation: Stated Importance vs. Actual Ratings")
plt.ylabel("Correlation Coefficient")
plt.xlabel("Trait")
plt.xticks(rotation=45)
plt.tight_layout()

correlation_series.sort_values(ascending=False)
```

- The correlations between the attributes are very low, suggesting that what people say they value doesn't align strongly with how they rate their dates
- "Fun" and "Attractiveness" had the highest (though still weak) alignment between preference and rating.
- Surprisingly, intelligence, ambition, and shared interests show negative correlations

Let's dive into whether shared interests or shared racial background are more influential in getting a match.

**Shared interests vs. shared racial background**

```python
# Shared interests vs. shared racial background
# 'match' = 1 if there was mutual interest
# 'samerace' = 1 if partners were of same race
# 'int_corr' = interest correlation (measure of shared interests)

# Clean up and drop NaNs relevant to this analysis
match_data = df[['match', 'samerace', 'int_corr']].dropna()

# Group by whether there's a match and calculate averages
grouped_means = match_data.groupby('match').mean()

# Plotting
grouped_means.plot(kind='bar', figsize=(8, 6), title="Shared Interests and Race by Match Outcome")
plt.ylabel("Average Value")
plt.xlabel("Match (0 = No, 1 = Yes)")
plt.xticks(rotation=0)
plt.tight_layout()

grouped_means
```

- People who matched were slightly more likely to share the same race, but the difference is small.
- Shared interests had a slightly stronger impact than racial background.
- Overall, neither factor had a dramatic effect, but shared interests edge out race as a predictor of a mutual match.

Now let's explore if being someone's first or last date of the night affects the chance of a second date.

**Speed dating order**

```python
# We use the 'order' column to determine the order of the date during the night
# Lower values = earlier in the evening, higher = later
# Let's analyze the average match rate based on order buckets (e.g., early, middle, late)

# Create buckets: early (1-3), middle (4-6), late (7+)
df_order = df[['match', 'order']].dropna()
df_order['order_bucket'] = pd.cut(df_order['order'], bins=[0, 3, 6, df_order['order'].max()], labels=['Early', 'Middle', 'Late'])

# Calculate match rate per bucket
order_match_rate = df_order.groupby('order_bucket')['match'].mean()

# Plotting
order_match_rate.plot(kind='bar', figsize=(7, 5), title="Match Rate by Time of Night (Order of Date)")
plt.ylabel("Match Rate")
plt.xlabel("Order Bucket")
plt.xticks(rotation=0)
plt.tight_layout()

order_match_rate
```

- Earlier dates have the highest match rate, with the likelihood of matching decreasing as the night progresses.
- This could be due to fatigue, comparison bias, or simply better energy early on.

Let's check if people are good at predicting how desirable they are  like whether their self-perception matches how others rate them.

**Comparison of participants' self-ratings to how their partners rated them**

```python
# We'll compare participants' self-ratings to how their partners rated them
# Use 'attr', 'sinc', 'intel', 'fun', 'amb' (how they see themselves)
# Compare to 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o' (how their date rated them)

self_cols = ['attr', 'sinc', 'intel', 'fun', 'amb']
partner_cols = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o']
trait_labels = ['Attractiveness', 'Sincerity', 'Intelligence', 'Fun', 'Ambition']

# Filter out rows with missing data
self_vs_partner = df[self_cols + partner_cols].dropna()

# Compute correlation between self-perception and partner ratings
self_partner_corr = {
    trait: self_vs_partner[self] .corr(self_vs_partner[partner])
    for trait, self, partner in zip(trait_labels, self_cols, partner_cols)
}

# Convert to Series
self_partner_corr_series = pd.Series(self_partner_corr)

# Plotting
self_partner_corr_series.plot(kind='bar', figsize=(10, 6), title="Correlation: Self-Perception vs. Partner's Ratings")
plt.ylabel("Correlation Coefficient")
plt.xlabel("Trait")
plt.xticks(rotation=45)
plt.tight_layout()

self_partner_corr_series.sort_values(ascending=False)
```

- People are somewhat accurate in knowing if they're fun, it's the most aligned trait.
- Attractiveness had a negative correlation, suggesting some overestimation or mismatch in how people see their own looks vs. how others do.
- Overall, the correlations are low, meaning people aren't great at predicting how desirable they are, at the lest in the eyes of their dates.
